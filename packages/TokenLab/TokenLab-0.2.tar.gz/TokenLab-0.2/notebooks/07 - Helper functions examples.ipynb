# Helper functions

TokenLab supports various helper functions which can be used to define shapes of entities like user growth.

```python
import numpy as np
import scipy
import os
import sys
tokenlab_path=os.path.abspath("").replace('notebooks','src')
sys.path.insert(0,tokenlab_path)

import TokenLab
from TokenLab.utils.helpers import *
from matplotlib import pyplot as plt
```

```python
START=1
STOP=1000
NUM=500
```

## Logarithmic Saturated space



```python
space1=log_saturated_space(start=START,stop=STOP,num=NUM)
plt.plot(space1)
```

## Logistic space

```python
space2=logistic_saturated_space(start=START,stop=STOP,num=NUM,steepness=2)
plt.plot(space2)
```

```python
original_space=np.linspace(start=START,stop=STOP,num=60)/STOP
subtractor=np.linspace(start=-6,stop=6,num=60)

spaces=scipy.special.expit(3*(original_space+subtractor+2))
spaces_weights=spaces/max(spaces)
final=spaces_weights*(STOP+START)
```

```python
plt.plot(final)
```
